# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

Both graphs have advantages and disadvantages in the way they present their data. It is easier to extract individual data about the trends for each week from the first graph. It is very hard to compare the two weeks, however, which is where the second graph comes in. The second graph allows us to compare each day of the week and the trends associated with the entire week to compare. From teh first graph, the two weeks look very similar except the second week was busier later in the week. When we compare the weeks in the second graph, however, we see the weeks are actually very different both in the number of visitors and the busy days.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
#graph from pg 191 
data = {
    'Product A' : [395, 410, 415, 390, 300, 270, 260],
    'Product B' : [360, 400, 405, 380, 280, 260 , 250]
}

labels = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7']
x = np.arange(len(labels)) 

width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, data['Product A'], width, label='Product A', color='grey')
plt.bar(x + width/2, data['Product B'], width, label='Product B', color='black')

plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Daily Sales Comparison: Product A vs Product B')
plt.xticks(x, labels)
plt.legend()
plt.tight_layout()
plt.show()
```

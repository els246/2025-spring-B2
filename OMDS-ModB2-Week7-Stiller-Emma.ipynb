# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Preattentive attributes are specific features that accentuate certain data characteristics in a strategic way. Preattentive attributes can be features like size, color, and position, all of which can draw a viewers attention towards a specific set of data for a specific purpose. For this case, we want to viewers' eye to be drawn to the most recent 3 months of data. To do this, we must change that section of data in some way to draw the viewers attention there first. We can change the color to make that section of data stand out more, for example make the whole line color grey and change the past 3 months section to bright orange. We can change the shape of the line to little boxes during the past 3 months instead of the plain line. Or we could even change the size of the line to be much thicker where we want the viewer to focus. Finally, we can put data labels on the data points we really want to highlight. We can even change all of these things if we really want that section of graph to stand out.

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
```

```python
#Graph on page 115 of Storytelling with Data 
x_values = ['Jan', 'Feb', 'Mar', 'April', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y_values_blue = [160,170,250,140,200,150,130,202,160,139,149,177]
y_values_red = [160,170,250,140,200,140,110,156,126,104,124,140]

plt.plot(x_values, y_values_blue, label='Ticket Volume Received', color='gray', alpha=0.5)
plt.plot(x_values, y_values_red, label='Ticket Volume Processed', color='darkblue', alpha=0.5)

plt.ylim(0,300)
plt.title('Ticket Volume Processed and Received')
plt.xlabel('Month')
plt.ylabel('Tickets')
plt.legend()

x_label = ['Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y_label_gray = [202,160,139,149,177]
y_label_darkblue = [156,126,104,124,140]
plt.scatter(x_label, y_label_gray, color='gray', zorder=5 )
plt.scatter(x_label, y_label_darkblue, color='darkblue', zorder=5)

for i, label in enumerate(x_label):
    plt.text(x_label[i], y_label_gray[i], f'{y_label_gray[i]}', fontsize=8, ha='left', color='gray')

for i, label in enumerate(x_label):
    plt.text(x_label[i], y_label_darkblue[i], f'{y_label_darkblue[i]}', fontsize=8, ha='left', color='darkblue')
```

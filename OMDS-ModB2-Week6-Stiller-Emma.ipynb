# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

Advantages to Area Plot: The area plot presents a clear visual representation of the two different chemical concentrations. It is easy to see how the two different chemicals fluctuate individually and also how they might correlate with each other. Area plots are great if we want to see how two different trends compare with each other. 


Disadvantages to Area Plot: The area plot has too much data so it is beginning to become unclear and unspecific. It is hard to get exact data points from the graph. If our goal was to find exact chemical concentrations or dates then the separate line graphs would be much easier to read and extract precise data from. 


Changes to the Area Plot: The area plot has great color contrast making it easy to see the difference in the two graphs. The graph needs a y-axis label to describe what the y-axis represents and the quantities measured in the graph. The x-axis is labeled properly, however using the number to represent the month might be more clear if more months were introduced to the graph.

The dataset I will be focusing on today is a stroke prediction dataset. The dataset analyzes different lifestyle features such as age, bmi, hypertension, residence type, work type, smoking status, etc and their correlation with the occurence of a stroke. This data will hopefully be able to be implemented and used as a preventative method to prevent stroke occurence in the overall human population. 

```python
import matplotlib.pyplot as plt
```

```python
#correlation matrix to determine correlation between features 
df = pd.read_csv('stroke-prediction-dataset.zip')
#print(df)
df.dtypes
#use label encoder to preprocess categorical columns 
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in df.select_dtypes(include=['object']).columns:
    df[column] = label_encoder.fit_transform(df[column])

#plot the correlation matrix with a heatmap 
corr = df.corr()
print(corr)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm',fmt='.2f', linewidths=0.5)

```

From this heat matrix, we cand deduce that the 'ever_married' column and 'age' are highly correlated features as well as the 'age' and 'stroke' features. Additionally 'heart disease', 'hypertension', and 'average glucose level' are all correlated with 'age' which reflects the degradation in our health as we age. 

```python
# plot 'ever married' and 'age' since they have a large correlation

plt.scatter(df['age'], df['bmi'], alpha=0.4)
plt.ylabel('bmi')
plt.xlabel('age (years)')
plt.legend()
plt.title('Age vs BMI in Stroke Prediction')
```

```python
#pairplot of the entire dataset with reference to age

sns.pairplot(df, hue='age')
```

```python
# pairplot of specific columns in dataset 

sns.pairplot(df[['age', 'bmi', 'avg_glucose_level']])
```

Conclusions: 

While this data has very clear and useful data for predicting stroke occurence, many features of the dataset are binary which makes it more challenging to visualize them graphically. The features with the highest correlation with the occurence of a stroke were age, heart disease, and average glucose level. This makes sense from a medical point of view. As we age, our blood vessels lose their ability to flex which can increase the potential for a blockage to occur. Patients with heart disease already struggle with their circulatory and vasculatory systems which can be conducive to a blockages including strokes. If our blood glucose level is high, our cells swell which can result in an increased likelihood of a blockage and potentially a stroke. Each of these features has been individually proven to increase the risk of stroke and this data set emphasizes these findings. I would argue that diet is a confounding variable in this dataset. Diet influences our blood glucose, hypertension, heart disease diagnoses, and bmi. All of these features are correlated with this confouding variable. Diet is hard to measure quantitatively, however, which may be why it was not included in the dataset.
